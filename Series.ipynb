Pandas Series: A series is a one dimensional labeled array capable of holding any data type.The axis labels are collectively called the index 

<h1>Creating Series</h1>
There are multiple way to create  a pandas Series:

```python
import numpy as np
import pandas as pd
```

```python
labels =['a','b','c']
my_list = [12, 20, 30]
arr = np.array([10,20,30])
d={1:10, 2:20, 3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
