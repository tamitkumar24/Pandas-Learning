```python
import numpy as np
import pandas as pd
```

<h3> Finding Missing Data</h3>

```python
#np.nan means it have NULL value
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()#is there is any null value if yes then return True and otherwise False
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

<h3>Removing Missing Data</h3>

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)#thresh means you must have to contain that much amounnt of ono null value that satisfy
```

<h3>Filling The Missing Data</h3>

```python
df.fillna(0)#those value have null value which now contain 0 value
```

```python
df#All change i made are not seen it df because we doesnot apply inplace
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
#different value are place that different null
```

```python
df.fillna(df.mean())#filling null value with means value of their respected column
```
