```python
import  numpy as numpy
import pandas as pd
```

<h3>Creating a Dataframe</h3> 

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df=pd.DataFrame(data)
df
```

```python
#list
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
```

```python
#but if you want to print column name also then you should perform this 
columns= ["Name", "Age", "City", "Salary"]
df2= pd.DataFrame(data_list, columns=columns)
df2
```

<h3>Selecting and Indexing of columns</h3>

<h3>Creating a new Column.</h3>

```python

df2["Designation"] = ["Doctor","Eng.","Doctor","Eng."]
```

```python
df2
```

<h3>Remove Column</h3>

```python
#Axis 0 means horizontal and 1 means vertical
df2.drop("City", axis=1)
```

```python
df2.drop("Designation", axis=1)
#drop only remove and show copy of frame not originally delete original frame
```

```python
df2
#but it show all column insist of droping it 
```

```python
#what can you do , it can delete from main data also 
df2.drop("Designation", axis=1, inplace = True)
```

```python
df2
#now you can see data designation is originally delete from main frame
```

```python
#we can also drop row , so we see as 
df2.drop(0,axis= 0)
```

```python
#but it doesnot originally delete
df2
```

<h3> Conditional Selection</h3>

```python
df
```

```python
#I only want to see those people whose age is above 30
df[df["Age"]> 30]
```

```python
#I only want poeple whose age is above 30 and their city must be paris
df[(df["Age"]>30) & (df["City"] == 'Paris')]
```
